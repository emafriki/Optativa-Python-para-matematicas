## Ejercicio



Usando Python, encuentra el exponente $n$ más pequeño de tal manera que el número complejo $(3+4i)^{n}$ tenga módulo mayor que 100.



```python
abs(3+4j)
```

```python
x=3+4j
n=0
for i in range(1,100):
  mod=abs(x**i)
  if i==1:
     n=mod
     e=i
  if mod>=100:
     e=i
     break
print('El n más pequeño es:',e)
```
